```python
from vnpy.gateway.bitmex import BitmexGateway
from vnpy.event import EventEngine

from vnpy.trader.engine import MainEngine
from vnpy.trader.ui import create_qapp
from vnpy.trader.object import HistoryRequest
from vnpy.trader.constant import Interval
from vnpy.trader.utility import extract_vt_symbol

from datetime import datetime

import requests
```

```python
proxies = {
  "http": "socks5://127.0.0.1:1080",
  "https": "socks5://127.0.0.1:1080",
}

r = requests.get('https://www.bitmex.com/api/v1/instrument/active', proxies=proxies)
[d['symbol'] for d in r.json()]
```

```python
qapp = create_qapp()

event_engine = EventEngine()

main_engine = MainEngine(event_engine)

main_engine.add_gateway(BitmexGateway)
```

```python
vt_symbol = 'XRPU19.BITMEX'
symbol, exchange = extract_vt_symbol(vt_symbol)
contract = main_engine.get_contract(vt_symbol)

req = HistoryRequest(
    symbol=symbol,
    exchange=exchange,
    interval=Interval('1m'),
    start=datetime(2019, 1, 1),
    end=datetime(2019, 9, 6)
)
history = main_engine.query_history(req, contract.gateway_name)
```
